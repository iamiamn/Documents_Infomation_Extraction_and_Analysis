```python
y_idxs = []
if start_year > 80:
    start_year = start_year - 100
for i in range(len(years_80216)):
    year = years_80216[i]
    if year <= end_year and year >= start_year:
        y_idxs.append(i)
print("%d / %d files are from the choosen years" % ( len(y_idxs), len(years_80216)))
# print([i for i in range(len(years_80216)) if i not in y_idxs])
```
